```python
# ----------------------------
# 0) Setup
# ----------------------------
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error
import math
import random

# Reproducibility
SEED = 42
np.random.seed(SEED); random.seed(SEED); torch.manual_seed(SEED)
if torch.cuda.is_available(): torch.cuda.manual_seed_all(SEED)

device = 'cuda' if torch.cuda.is_available() else 'cpu'

# Assumes df already loaded with columns listed by you and 'state' present
# df columns (from your message):
# ['date','tavg','tmin','tmax','prcp','wspd','pres','state',
#  'total_vaccinations','total_distributed','people_vaccinated',
#  'people_fully_vaccinated_per_hundred','total_vaccinations_per_hundred',
#  'people_fully_vaccinated','people_vaccinated_per_hundred',
#  'distributed_per_hundred','daily_vaccinations_raw','daily_vaccinations',
#  'daily_vaccinations_per_million','share_doses_used','total_boosters',
#  'total_boosters_per_hundred','retail_and_recreation_percent_change_from_baseline',
#  'grocery_and_pharmacy_percent_change_from_baseline','parks_percent_change_from_baseline',
#  'transit_stations_percent_change_from_baseline','workplaces_percent_change_from_baseline',
#  'residential_percent_change_from_baseline','fips','cases','deaths','growth_rate']

# ----------------------------
# 1) Feature groups + target
# ----------------------------
environment_features = ['tavg', 'tmin', 'tmax', 'prcp', 'wspd', 'pres']

mobility_features = [
    'retail_and_recreation_percent_change_from_baseline',
    'grocery_and_pharmacy_percent_change_from_baseline',
    'parks_percent_change_from_baseline',
    'transit_stations_percent_change_from_baseline',
    'workplaces_percent_change_from_baseline',
    'residential_percent_change_from_baseline'
]

# Prefer per-hundred / per-million vaccination signals for scale stability
vaccine_features = [
    'people_fully_vaccinated_per_hundred',
    'total_vaccinations_per_hundred',
    'people_vaccinated_per_hundred',
    'distributed_per_hundred',
    'daily_vaccinations_per_million',
    'share_doses_used',
    'total_boosters_per_hundred'
]

# Temporal: include growth_rate predictors (AR-ish same-day) + cases/deaths levels
temporal_features = ['cases', 'deaths']  # we keep target separate

# Target
target_col = 'growth_rate'

# Minimal cleaning
df = df.dropna(subset=['state', target_col]).copy()
df['date'] = pd.to_datetime(df['date'])

# Encode states (for embeddings + LOSO splits)
le = LabelEncoder()
df['state_id'] = le.fit_transform(df['state'])

# ----------------------------
# 2) SNN model (module blocks + state embedding)
# ----------------------------
class SNNModel(nn.Module):
    def __init__(self, input_dims, state_count, emb_dim=8, hidden_dim=64, dropout=0.3):
        super().__init__()
        self.state_emb = nn.Embedding(state_count, emb_dim)

        def block(in_dim):
            return nn.Sequential(
                nn.Linear(in_dim, hidden_dim),
                nn.ReLU(),
                nn.Dropout(dropout)
            )

        self.temporal = block(input_dims['temporal'])
        self.mobility = block(input_dims['mobility'])
        self.vaccine = block(input_dims['vaccine'])
        self.environment = block(input_dims['environment'])

        total_in = hidden_dim * 4 + emb_dim
        self.fusion = nn.Sequential(
            nn.Linear(total_in, hidden_dim),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim, 1),
            nn.Tanh()   # keep outputs bounded; target is roughly z-scored-ish
        )

    def forward(self, x_modules, state_ids):
        tmp = self.temporal(x_modules['temporal'])
        mob = self.mobility(x_modules['mobility'])
        vac = self.vaccine(x_modules['vaccine'])
        env = self.environment(x_modules['environment'])
        emb = self.state_emb(state_ids)
        fused = torch.cat([tmp, mob, vac, env, emb], dim=1)
        out = self.fusion(fused)
        return out.squeeze()

    def forward_with_contributions(self, x_modules, state_ids):
        tmp = self.temporal(x_modules['temporal'])
        mob = self.mobility(x_modules['mobility'])
        vac = self.vaccine(x_modules['vaccine'])
        env = self.environment(x_modules['environment'])
        emb = self.state_emb(state_ids)
        fused = torch.cat([tmp, mob, vac, env, emb], dim=1)
        out = self.fusion(fused)
        return out.squeeze(), {'temporal': tmp, 'mobility': mob, 'vaccine': vac, 'environment': env}

# Input dims
input_dims = {
    'temporal': len(temporal_features),
    'mobility': len(mobility_features),
    'vaccine': len(vaccine_features),
    'environment': len(environment_features)
}

# ----------------------------
# 3) Helper: build tensors for any dataframe with fit scalers from train
# ----------------------------
def build_tensors(train_df, test_df):
    # Fit scalers on TRAIN ONLY (per module)
    scalers = {}
    def fit_transform(df_part, cols, key):
        sc = StandardScaler()
        vals = sc.fit_transform(df_part[cols].astype(float).values)
        scalers[key] = sc
        return vals

    def transform(df_part, cols, key):
        sc = scalers[key]
        return sc.transform(df_part[cols].astype(float).values)

    Xtr_env = fit_transform(train_df, environment_features, 'env')
    Xtr_mob = fit_transform(train_df, mobility_features, 'mob')
    Xtr_vac = fit_transform(train_df, vaccine_features, 'vac')
    Xtr_tmp = fit_transform(train_df, temporal_features, 'tmp')

    Xte_env = transform(test_df, environment_features, 'env')
    Xte_mob = transform(test_df, mobility_features, 'mob')
    Xte_vac = transform(test_df, vaccine_features, 'vac')
    Xte_tmp = transform(test_df, temporal_features, 'tmp')

    # Target: standardize using TRAIN stats (safer)
    y_sc = StandardScaler()
    ytr = y_sc.fit_transform(train_df[[target_col]].values).ravel()
    yte = y_sc.transform(test_df[[target_col]].values).ravel()

    state_tr = torch.tensor(train_df['state_id'].values, dtype=torch.long)
    state_te = torch.tensor(test_df['state_id'].values, dtype=torch.long)

    # Pack into tensors
    train_modules = {
        'temporal': torch.tensor(Xtr_tmp, dtype=torch.float32),
        'mobility': torch.tensor(Xtr_mob, dtype=torch.float32),
        'vaccine': torch.tensor(Xtr_vac, dtype=torch.float32),
        'environment': torch.tensor(Xtr_env, dtype=torch.float32)
    }
    test_modules = {
        'temporal': torch.tensor(Xte_tmp, dtype=torch.float32),
        'mobility': torch.tensor(Xte_mob, dtype=torch.float32),
        'vaccine': torch.tensor(Xte_vac, dtype=torch.float32),
        'environment': torch.tensor(Xte_env, dtype=torch.float32)
    }
    y_train_t = torch.tensor(ytr, dtype=torch.float32)
    y_test_t  = torch.tensor(yte, dtype=torch.float32)

    return train_modules, test_modules, state_tr, state_te, y_train_t, y_test_t, y_sc

# ----------------------------
# 4) LOSO training + explainability
# ----------------------------
states = df['state'].unique().tolist()
results = []

for held_out in states:
    train_df = df[df['state'] != held_out].copy()
    test_df  = df[df['state'] == held_out].copy()

    # Skip tiny test sets
    if len(test_df) < 50 or len(train_df) < 500:
        continue

    # Build tensors with per-fold scaling
    Xtr_mods, Xte_mods, state_tr, state_te, y_tr, y_te, y_scaler = build_tensors(train_df, test_df)

    # Dataloaders
    train_ds = TensorDataset(Xtr_mods['temporal'], Xtr_mods['mobility'], Xtr_mods['vaccine'], Xtr_mods['environment'], state_tr, y_tr)
    test_ds  = TensorDataset(Xte_mods['temporal'], Xte_mods['mobility'], Xte_mods['vaccine'], Xte_mods['environment'], state_te, y_te)
    train_dl = DataLoader(train_ds, batch_size=256, shuffle=True)
    val_dl   = DataLoader(test_ds, batch_size=512, shuffle=False)

    # Model
    model = SNNModel(input_dims, state_count=df['state_id'].nunique(), emb_dim=8, hidden_dim=64, dropout=0.3).to(device)
    criterion = nn.MSELoss()
    optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-4)

    # Train w/ early stopping
    best_val, patience, wait = float('inf'), 8, 0
    for epoch in range(80):
        model.train()
        train_loss = 0.0
        for tmp, mob, vac, env, st, y in train_dl:
            tmp, mob, vac, env, st, y = tmp.to(device), mob.to(device), vac.to(device), env.to(device), st.to(device), y.to(device)
            optimizer.zero_grad()
            preds = model({'temporal': tmp, 'mobility': mob, 'vaccine': vac, 'environment': env}, st)
            loss = criterion(preds, y)
            loss.backward()
            optimizer.step()
            train_loss += loss.item() * len(y)
        train_loss /= len(train_dl.dataset)

        model.eval()
        val_loss, preds_list, trues_list = 0.0, [], []
        with torch.no_grad():
            for tmp, mob, vac, env, st, y in val_dl:
                tmp, mob, vac, env, st, y = tmp.to(device), mob.to(device), vac.to(device), env.to(device), st.to(device), y.to(device)
                out = model({'temporal': tmp, 'mobility': mob, 'vaccine': vac, 'environment': env}, st)
                loss = criterion(out, y)
                val_loss += loss.item() * len(y)
                preds_list.extend(out.cpu().numpy()); trues_list.extend(y.cpu().numpy())
        val_loss /= len(val_dl.dataset)

        # simple early stop
        if val_loss < best_val:
            best_val, wait = val_loss, 0
            best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}
        else:
            wait += 1
            if wait >= patience:
                break

    # Load best
    model.load_state_dict(best_state)

    # Predict + metrics (on standardized target)
    model.eval()
    with torch.no_grad():
        tmp = Xte_mods['temporal'].to(device)
        mob = Xte_mods['mobility'].to(device)
        vac = Xte_mods['vaccine'].to(device)
        env = Xte_mods['environment'].to(device)
        st  = state_te.to(device)
        out_std, acts = model.forward_with_contributions({'temporal': tmp, 'mobility': mob, 'vaccine': vac, 'environment': env}, st)

    y_true_std = y_te.cpu().numpy()
    y_pred_std = out_std.cpu().numpy()

    # MAE/RMSE on standardized scale ( unitless). If you want original scale, invert with y_scaler.
    mae = mean_absolute_error(y_true_std, y_pred_std)
    rmse = math.sqrt(mean_squared_error(y_true_std, y_pred_std))

    contrib_scores = {name: acts[name].abs().mean().item() for name in acts}

    results.append({
        'held_out_state': held_out,
        'MAE_std': mae,
        'RMSE_std': rmse,
        'contrib_temporal': contrib_scores['temporal'],
        'contrib_mobility': contrib_scores['mobility'],
        'contrib_vaccine': contrib_scores['vaccine'],
        'contrib_environment': contrib_scores['environment'],
    })

# Results table
loso_results = pd.DataFrame(results).sort_values('RMSE_std')
display(loso_results.head(10))
display(loso_results.tail(10))

# Optional: normalize contributions per state so rows sum to 1 (easier to compare)
norm_cols = ['contrib_temporal','contrib_mobility','contrib_vaccine','contrib_environment']
norm = loso_results[norm_cols].div(loso_results[norm_cols].sum(axis=1), axis=0)
norm['held_out_state'] = loso_results['held_out_state']
norm = norm[['held_out_state'] + norm_cols]
display(norm.sort_values('contrib_temporal', ascending=False).head(10))
```

```python
import numpy as np

def create_sequences(df, feature_cols, target_col, seq_len=14, horizon=7):
    X, y = [], []
    states = df['state_id'].unique()

    for state in states:
        state_df = df[df['state_id'] == state].sort_values('date')
        features = state_df[feature_cols].values
        target = state_df[target_col].values

        for i in range(len(state_df) - seq_len - horizon):
            X.append(features[i:i+seq_len])
            y.append(target[i+seq_len + horizon - 1])

    return np.array(X), np.array(y)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# --- 1 Model Definition ---
class SNNModel(nn.Module):
    def __init__(self, input_dims, state_count, emb_dim=8, hidden_dim=64, dropout=0.3):
        super(SNNModel, self).__init__()
        self.state_emb = nn.Embedding(state_count, emb_dim)

        def block(in_dim):
            return nn.Sequential(
                nn.Linear(in_dim, hidden_dim),
                nn.ReLU(),
                nn.Dropout(dropout)
            )

        self.temporal = block(input_dims['temporal'])
        self.mobility = block(input_dims['mobility'])
        self.vaccine = block(input_dims['vaccine'])
        self.environment = block(input_dims['environment'])

        total_in = hidden_dim * 4 + emb_dim
        self.fusion = nn.Sequential(
            nn.Linear(total_in, hidden_dim),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim, 1),
            nn.Tanh()
        )

    def forward(self, x_modules, state_ids):
        tmp = self.temporal(x_modules['temporal'])
        mob = self.mobility(x_modules['mobility'])
        vac = self.vaccine(x_modules['vaccine'])
        env = self.environment(x_modules['environment'])
        emb = self.state_emb(state_ids)
        fused = torch.cat([tmp, mob, vac, env, emb], dim=1)
        out = self.fusion(fused)
        return out.squeeze()


# --- 2 Input dimensions ---
input_dims = {
    'temporal': 12,
    'mobility': 12,
    'vaccine': 5,
    'environment': 12
}

# --- 3 Prepare data ---
device = 'cuda' if torch.cuda.is_available() else 'cpu'

X_train_tensors = {k: torch.tensor(v, dtype=torch.float32) for k, v in X_train_modules.items()}
X_test_tensors = {k: torch.tensor(v, dtype=torch.float32) for k, v in X_test_modules.items()}

y_train_t = torch.tensor(y_train, dtype=torch.float32)
y_test_t = torch.tensor(y_test, dtype=torch.float32)
state_train_t = torch.tensor(state_ids_train, dtype=torch.long)
state_test_t = torch.tensor(state_ids_test, dtype=torch.long)

# Move to device later in loop for efficiency
train_ds = TensorDataset(
    X_train_tensors['temporal'], X_train_tensors['mobility'],
    X_train_tensors['vaccine'], X_train_tensors['environment'],
    state_train_t, y_train_t
)
test_ds = TensorDataset(
    X_test_tensors['temporal'], X_test_tensors['mobility'],
    X_test_tensors['vaccine'], X_test_tensors['environment'],
    state_test_t, y_test_t
)

train_dl = DataLoader(train_ds, batch_size=256, shuffle=True)
val_dl = DataLoader(test_ds, batch_size=512, shuffle=False)


# --- 4 Initialize Model, Optimizer, Criterion ---
model = SNNModel(input_dims, state_count=len(df['state_id'].unique())).to(device)
criterion = nn.MSELoss()
optimizer = optim.AdamW(model.parameters(), lr=1e-5, weight_decay=1e-4)

# --- 5 Training Loop with Early Stopping ---
best_val = float('inf')
patience, wait = 10, 0
epochs = 150

for epoch in range(epochs):
    model.train()
    train_loss = 0.0
    for tmp, mob, vac, env, state, y in train_dl:
        tmp, mob, vac, env, state, y = tmp.to(device), mob.to(device), vac.to(device), env.to(device), state.to(device), y.to(device)
        optimizer.zero_grad()
        outputs = model({'temporal': tmp, 'mobility': mob, 'vaccine': vac, 'environment': env}, state)
        loss = criterion(outputs, y)
        loss.backward()
        optimizer.step()
        train_loss += loss.item() * len(y)

    train_loss /= len(train_dl.dataset)

    # Validation
    model.eval()
    val_loss = 0.0
    preds, trues = [], []
    with torch.no_grad():
        for tmp, mob, vac, env, state, y in val_dl:
            tmp, mob, vac, env, state, y = tmp.to(device), mob.to(device), vac.to(device), env.to(device), state.to(device), y.to(device)
            out = model({'temporal': tmp, 'mobility': mob, 'vaccine': vac, 'environment': env}, state)
            loss = criterion(out, y)
            val_loss += loss.item() * len(y)
            preds.extend(out.cpu().numpy())
            trues.extend(y.cpu().numpy())

    val_loss /= len(val_dl.dataset)

    print(f"Epoch {epoch:03d} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")

    if val_loss < best_val:
        best_val, wait = val_loss, 0
        torch.save(model.state_dict(), 'best_snn_model.pt')
    else:
        wait += 1
        if wait >= patience:
            print("Early stopping triggered.")
            break


# --- 6 Evaluate on Validation Set ---
model.load_state_dict(torch.load('best_snn_model.pt'))
model.eval()
preds = np.array(preds)
trues = np.array(trues)
mae = mean_absolute_error(trues, preds)
rmse = np.sqrt(mean_squared_error(trues, preds))
print(f"\nFinal Evaluation -> MAE: {mae:.4f} | RMSE: {rmse:.4f}")
```

```python
# --- 1 Updated Model Definition (reduced temporal capacity + activation outputs) ---
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

class SNNModel(nn.Module):
    def __init__(self, input_dims, state_count, emb_dim=8, hidden_dim=64, dropout=0.3):
        super(SNNModel, self).__init__()
        self.state_emb = nn.Embedding(state_count, emb_dim)

        def block(in_dim, out_dim=None):
            out_dim = out_dim or hidden_dim
            return nn.Sequential(
                nn.Linear(in_dim, out_dim),
                nn.ReLU(),
                nn.Dropout(dropout)
            )

        #  smaller capacity for temporal module
        self.temporal = block(input_dims['temporal'], hidden_dim // 2)
        self.mobility = block(input_dims['mobility'], hidden_dim)
        self.vaccine = block(input_dims['vaccine'], hidden_dim)
        self.environment = block(input_dims['environment'], hidden_dim)

        total_in = (hidden_dim // 2) + hidden_dim * 3 + emb_dim
        self.fusion = nn.Sequential(
            nn.Linear(total_in, hidden_dim),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim, 1),
            nn.Tanh()
        )

    def forward_with_acts(self, x_modules, state_ids):
        tmp = self.temporal(x_modules['temporal'])
        mob = self.mobility(x_modules['mobility'])
        vac = self.vaccine(x_modules['vaccine'])
        env = self.environment(x_modules['environment'])
        emb = self.state_emb(state_ids)
        fused = torch.cat([tmp, mob, vac, env, emb], dim=1)
        out = self.fusion(fused)
        return out.squeeze(), {'temporal': tmp, 'mobility': mob, 'vaccine': vac, 'environment': env}


# --- 2 Training Setup ---
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = SNNModel(input_dims, state_count=len(df['state_id'].unique())).to(device)

criterion = nn.MSELoss()
optimizer = optim.AdamW(model.parameters(), lr=1e-5, weight_decay=1e-4)

train_ds = TensorDataset(
    X_train_tensors['temporal'], X_train_tensors['mobility'],
    X_train_tensors['vaccine'], X_train_tensors['environment'],
    state_train_t, y_train_t
)
test_ds = TensorDataset(
    X_test_tensors['temporal'], X_test_tensors['mobility'],
    X_test_tensors['vaccine'], X_test_tensors['environment'],
    state_test_t, y_test_t
)
train_dl = DataLoader(train_ds, batch_size=256, shuffle=True)
val_dl = DataLoader(test_ds, batch_size=512, shuffle=False)


# --- 3 Training Loop with Explicit Regularization ---
best_val, patience, wait = float('inf'), 10, 0
epochs = 150

for epoch in range(epochs):
    model.train()
    train_loss = 0.0
    for tmp, mob, vac, env, state, y in train_dl:
        tmp, mob, vac, env, state, y = tmp.to(device), mob.to(device), vac.to(device), env.to(device), state.to(device), y.to(device)
        optimizer.zero_grad()

        outputs, acts = model.forward_with_acts({'temporal': tmp, 'mobility': mob, 'vaccine': vac, 'environment': env}, state)
        loss = criterion(outputs, y)

        #  regularization term to encourage balanced module use
        mean_t = acts['temporal'].abs().mean()
        mean_m = acts['mobility'].abs().mean()
        mean_v = acts['vaccine'].abs().mean()
        mean_e = acts['environment'].abs().mean()
        reg = (mean_t - (mean_m + mean_v + mean_e) / 3) ** 2
        loss = loss + 1e-3 * reg

        loss.backward()
        optimizer.step()
        train_loss += loss.item() * len(y)
    train_loss /= len(train_dl.dataset)

    # Validation
    model.eval()
    val_loss, preds, trues = 0.0, [], []
    with torch.no_grad():
        for tmp, mob, vac, env, state, y in val_dl:
            tmp, mob, vac, env, state, y = tmp.to(device), mob.to(device), vac.to(device), env.to(device), state.to(device), y.to(device)
            out, _ = model.forward_with_acts({'temporal': tmp, 'mobility': mob, 'vaccine': vac, 'environment': env}, state)
            loss = criterion(out, y)
            val_loss += loss.item() * len(y)
            preds.extend(out.cpu().numpy())
            trues.extend(y.cpu().numpy())
    val_loss /= len(val_dl.dataset)

    print(f"Epoch {epoch:03d} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")

    if val_loss < best_val:
        best_val, wait = val_loss, 0
        torch.save(model.state_dict(), 'balanced_snn.pt')
    else:
        wait += 1
        if wait >= patience:
            print("Early stopping triggered.")
            break


# --- 4 Evaluate ---
model.load_state_dict(torch.load('balanced_snn.pt', map_location=device))
model.eval()
mae = mean_absolute_error(trues, preds)
rmse = np.sqrt(mean_squared_error(trues, preds))
print(f"\nFinal Evaluation -> MAE: {mae:.4f} | RMSE: {rmse:.4f}")

# --- 5 Contribution Visualization ---
with torch.no_grad():
    _, acts = model.forward_with_acts(X_test_tensors, state_test_t.to(device))

contrib_scores = {name: act.abs().mean().item() for name, act in acts.items()}

plt.figure(figsize=(6,4))
plt.bar(contrib_scores.keys(), contrib_scores.values(),
        color=['royalblue','darkorange','seagreen','purple'])
plt.title("Average Module Contribution Magnitude (Balanced SNN)")
plt.ylabel("Mean |Activation|")
plt.grid(alpha=0.3)
plt.show()

print("Contribution scores:", contrib_scores)
```
